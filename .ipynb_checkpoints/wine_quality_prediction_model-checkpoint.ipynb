# Import data

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from ucimlrepo import fetch_ucirepo

# fetch dataset and convert to pandas dataframe
wine_quality = fetch_ucirepo(id=186)
df = pd.concat([wine_quality.data.features, wine_quality.data.targets], axis=1)

# Add the 'color' column from original dataset which is not included in features
df['color'] = wine_quality.data.original['color']

# Redefine df to only include red wines
df = df[df['color'] == 'red']

# drop color as it is no longer needed
df = df.drop('color', axis=1)

print(df.head())

X = df.drop('quality', axis=1)
y = df['quality']
```

# Create plot 1: histogram

```python
# histogram
plt.figure(figsize=(8, 6))
sns.histplot(df['quality'], bins=6, color='#7A0422')
plt.title('Distribution of Red Wine Quality')
plt.xlabel('Quality Score')
plt.ylabel('Frequency')
plt.show()
```

# Create plot 2: Heatmap

```python
plt.figure(figsize=(12, 10))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Red Wine Features')
plt.show()
```

# Plot 3: scatterplots?

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='alcohol', y='quality', data=df)
plt.title('Relationship between alcohol and quality')
plt.xlabel('alcohol')
plt.ylabel('quality')
plt.show()

plt.figure(figsize=(8, 6))
sns.scatterplot(x='alcohol', y='density', data=df, color='#7A0422')
plt.title('Relationship between alcohol and density')
plt.xlabel('alcohol')
plt.ylabel('density')
plt.show()

plt.figure(figsize=(8, 6))
sns.scatterplot(x='fixed_acidity', y='density', data=df)
plt.title('Relationship between f-Acidity and density')
plt.xlabel('fixed_acidity')
plt.ylabel('density')
plt.show()
```

# box plots

```python
# Example: Box plots of quality by alcohol content (binned)
df['alcohol_bin'] = pd.cut(df['alcohol'], bins=5)  # Create 5 bins for alcohol

plt.figure(figsize=(10, 6))
sns.boxplot(x='alcohol_bin', y='quality', data=df)
plt.title('Box Plots of Quality by Alcohol Content (Binned)')
plt.xlabel('Alcohol Content (Bins)')
plt.ylabel('Quality Score')
plt.show()

plt.figure(figsize=(10, 6))
sns.violinplot(x='alcohol_bin', y='quality', data=df)
plt.title('Violin Plots of Quality by Alcohol Content (Binned)')
plt.xlabel('Alcohol Content (Bins)')
plt.ylabel('Quality Score')
plt.show()


# # Example: Box plots of quality by volatile acidity (binned)
df['volatile_acidity_bin'] = pd.cut(df['volatile_acidity'], bins=5)  # Create 5 bins for volatile acidity

plt.figure(figsize=(10, 6))
sns.boxplot(x='volatile_acidity_bin', y='quality', data=df)
plt.title('Box Plots of Quality by volatile acidity Content (Binned)')
plt.xlabel('volatile acidity Content (Bins)')
plt.ylabel('Quality Score')
plt.show()

plt.figure(figsize=(10, 6))
sns.violinplot(x='volatile_acidity_bin', y='quality', data=df)
plt.title('Violin Plots of Quality by volatile acidity Content (Binned)')
plt.xlabel('volatile acidity Content (Bins)')
plt.ylabel('Quality Score')
plt.show()
```

# pair plots

```python
sns.pairplot(df, vars=['alcohol', 'volatile_acidity', 'sulphates', 'quality']) #vars is used to select specific columns, otherwise all columns are used.
plt.show()
```

```python
sns.kdeplot(df['alcohol'])
plt.title('Density Plot of Alcohol Content')
plt.xlabel('Alcohol Content')
plt.ylabel('Density')
plt.show()
```

```python
# Example: Bar chart of average quality by alcohol bin
average_quality_by_bin = df.groupby('alcohol_bin', observed=True)['quality'].mean().reset_index()
sns.barplot(x='alcohol_bin', y='quality', data=average_quality_by_bin)
plt.title('Average Quality by Alcohol Content (Binned)')
plt.xlabel('Alcohol Content (Bins)')
plt.ylabel('Average Quality')
plt.show()
```
